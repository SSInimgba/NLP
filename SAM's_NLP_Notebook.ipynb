## Question 2 ##

* Apply the following preprocessing steps:

     1. Remove all words that contain numbers
     2. Make all the text lowercase
     3. Remove punctuation
     4. Tokenize the reviews into words
     
  Hint #1: Use regular expressions.
  
  Hint #2: The cookie review in the second row has numbers, upper case letters and punctuation. You can use it to test out your regular expressions.
     
     
* Find the most common words.
* Determine the word length distribution over the entire corpus.

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
```

```python
data['reviews_preprocessed'] = data['reviews'].str.replace('\d+','') #remove number
data['reviews_preprocessed'] = data['reviews_preprocessed'].str.lower()  #make to lower case
data['reviews_preprocessed'] = data['reviews_preprocessed'].str.replace('<[^<]+?>',' ') #remove links
data['reviews_preprocessed']= data['reviews_preprocessed'].str.replace('[^\w\s]','') #remove punctuation
data['reviews_preprocessed'] = data['reviews_preprocessed'].apply(word_tokenize) #tokenize
data.head()
```

```python
from collections import Counter

counter = Counter()
_ = data['reviews_preprocessed'].apply(lambda x: counter.update(x))

counter.most_common(50) 
```

```python
#Determine the word length distribution over the entire corpus.
words = []
for i in data['reviews_preprocessed']:
  words.append(i)
words
```

```python
for i in range(len(words)):
  for x in words[i]:
    print(x,len(x))
```
